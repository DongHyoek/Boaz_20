# Chapter 1
## 한눈에 보는 머신러닝

#### 머신러닝이란?

- Training set을 가지고 시스템(모델)이 학습을 하는 것으로 **경험 E**를 통해 **새로운 작업 T**를 진행하고, **성능 측정 P**는 직접 정의해야한다. 이 떄 성능 측정을 accuracy(정확도)라고 부르며 흔히 classification task에서 자주 사용된다.


#### 왜 머신러닝을 사용해야 하는지?

- 만약 스팸 필터를 만들 떄 사람이 직접 작업을 한다면 각 패턴을 감지하는 알고리즘을 일일이 작성해야 하므로 **규칙이 매우 길어지고, 복잡해져 유지 보수가 어려워진다.**

  그러나 머신러닝에 기반한 스팸 필터는 자동으로 스팸 메일을 판단하는데에 좋은 기준을 자동으로 학습하여 **프로그램이 훨씬 짧아지고 유지 보수가 쉬우며, 대부분 정확도가 더 높다.**
  
  이렇게 머신러닝 기술을 접목시켜 데이터를 분석하게 되면 사람이 규칙을 정의하는 것 보다 겉으로는 보이지 않던 데이터의 패턴들을 발견할 수 있고, 대용량 데이터를 분석하면 이를 **데이터 마이닝**이라고 한다.
  
  정리하자면 머신러닝이란 **숨겨진 데이터의 패턴을 발견하는 것이다.**
  
  
####  머신러닝 시스템의 종류 

1) 지도 학습 (Supervised Learning)

   - 정답(label 혹은 class)이 있는 데이터로 모델을 학습시키는 것으로 분류, 회귀 문제에서 가장 많이 사용되는 학습 방법.
   
   ex) K-최근접이웃, 선형 회귀, 로지스틱 회귀(분류임), 서포트 벡터 머신, 결정트리 & 랜덤 포레스트, 신경망

2) 비지도 학습 (Unsupervised Learning)
    
   - 지도 학습과는 반대로 정답이 없는 데이터로 학습하는 학습방법으로 크게 3가지의 알고리즘들이 있다.
   
       1) 군집화 (Clustering)
       
           - k-means, DBSCAN, 계층군집분석(HCA), 이상치 탐지와 특이치 탐지, 원-클래스(SVM), 아이솔레이션 포레스트(isolation forest)
           
       2) 시각화와 차원축소 
       
          레이블이 없는 대규모의 고차원 데이터를 2D나 3D로 표현하므로 숨겨져있던 데이터의 패턴을 발견하는 머신러닝이라고 생각할 수 있다. 그리고 pca와 같은 방법들을 feature extraction이라고 한다.
           
           - pca, kernel pca, LLE(지역적 선형 임베딩), t-SNE
           
       3) 연관규칙 학습 (association rule learning)
       

       
            
           군집화와 비슷하게 느껴질 수도 있겠지만 약간의 차이점이 있으며, 흔히 장바구니 분석이라고도 부른다. 대부분의 추천시스템 등은 연관규칙 학습에 기반한 분석방법이다.
       
           - 어프라이어리, 이클렛
   
       
3) 준지도 학습 (Semisupervised Learning)

   - 데이터에 레이블을 다는 것은 일반적으로 시간과 비용이 많이 들기 때문에, 레이블이 있는 샘플보다 없는 샘플이 많은 경우가 많다. 그래서 일부만 레이블이 있는 데이터로 학습을 진행하는 방법으로 **대부분 지도 와 비지도의 조합으로 이루어져 있다**
    
    cf) 심층 신뢰 신경망(DBN)은 여러 겹의 제한된 볼츠만 머신(RBM)이라는 비지도 학습에 기초한다.
    


     예시로 위 그림과 같이 일부만 있는 클래스로 임의의 데이터가 어느 영역에 더 가까운지 보고 해당 클래스로 분류하는 것이 준지도 학습 방법. 

4) 강화학습(Reinforcement learning)



   - 기존과는 매우 다른 종류의 알고리즘으로 여기서는 학습하는 시스템을 **에이전트**라고 부르며 **지속적은 보상과 패널티를 적절하게 부여**하며 시간이 지나면서 모델이 스스로 가장 큰 보상을 얻기위한 **최상의 전략인 정책(Policy)을 스스로 학습한다.** 
    
    정책이란 주어진 상황에서 에이전트(모델)이 어떤 행동을 선택해야 하는지 정의하는 것.
    
    대표적인 예시로 구글의 알파고가 있다.
    
    
#### 배치 학습과 미니배치 학습

1) 배치 학습 

  - 가용한 데이터를 모두 사용해 훈련시키는 방법으로 시간과 자원을 많이 소모하므로 보통 오프라인에서 수행된다. 먼저 모델을 훈련시키고 그 다음 제품 시스템에 적용하면 더 이상의 학습 없이 실행되므로 이를 **오프라인 학습**이라고 한다.
    
    - 그러나 새로운 데이터에 대해 학습하려면 이전 데이터까지도 포함된 전체 데이터를 새롭게 재구성하여 처음부터 다시 학습해야 한다. -> 비용과 시간적 차원에서 큰 손실.
    

2) 미니배치 학습 

  - 데이터를 순차적으로 한 개 혹은 작은 묶음 단위인 **미니 배치** 데이터로 모델을 훈련시키는 방법이다. 이를 **온라인 학습 or 점진적 학습**이라고 하며 이러한 방법은 시간적, 자원적 요소들이 배치 학습보다 현저하게 적게 사용되므로 빠른 변화에 적응해야하는 시스템에 적합하다. 학습이 끝난 데이터는 더는 필요하지 않다(?)
  
    학습률을 통해 학습의 속도를 조정할 수 있음.

    - 그러나 만약 noise가 낀 데이터가 주입되었을 때 성능이 점진적으로 감소될 수 있으므로 이상치 탐지 알고리즘을 사용해서 모니터링하거나, 학습을 중지시켜야 한다.
    
#### 사례 기반 학습과 모델 기반 학습

- 머신러닝 시스템의 목표는 주어진 훈련 데이터로 학습하고 **새로운 test 데이터에서 일반화된 좋은 예측(성능)**을 만들어내야 하는 것이다. 이를 위한 2가지 접근 방법이 있다.

1) 사례 기반 학습

   - 이름과 같이 사례들을 기억하는 접근법으로 **유사도를 측정함으로써** 원하는 target과 매우 유사한 샘플들을 구분하도록 프로그래밍 하는 방법이다.

2) 모델 기반 학습

   - 이름과 같이 모델을 만들어서 예측에 사용하는 것이다.
   
       일반적인 과정으로는 모델 선택 ->  비용함수로 최적 파라미터 찾기 (모델학습) -> 새로운 데이터로 평가
  

#### 머신러닝의 주요 도전 과제

#####  데이터 관점

1) 충분하지 않은 양의 훈련 데이터 
  
  - 데이터 양이 많을수록 성능이 좋고, 복잡한 문제일수록 데이터가 많이 필요함.

2) 대표성 없는 훈련 데이터

  - 일반화를 위한 대표성이 없는 데이터를 사용했을 땐 잘못된 결과를 도출할 수 있음. 흔히 샘플이 작으면 noise(우연에 의한)가 생기게 되고, 매우 큰 샘플도 편향되게 잘못 추출한다면 잘못된 결과가 도출된다.

3) 낮은 품질의 데이터
    
  - 기본적은 raw data에는 이상치, noise 등 성능을 낮추는 요인들이 많이 포함되어있으므로, 이를 걸러내기 위한 **data preprocessing 과정**이 꼭 필요하다.
  
4) 관련 없는 특성

  - target과 관련된 feature들이 들어가야지 좋은 결과물이 도출되므로, **feature engineering 과정**이 꼭 필요하다.
  
##### 알고리즘 관점

5) 과대적합과 과소적합

   - 과대적합 
   
       훈련 데이터에 과도하게 적합되어 새로운 데이터에서는 성능이 크게 떨어지는 일반화 성능이 떨어지는 상황. 보통 이 경우에는 훈련 데이터에 있는 noise의 양에 비해 **모델이 너무 복잡할 때** 일어난다. 
       
       cf) 이를 위해 모델의 단순화를 위한 하이퍼파라미터를 사용해 규제양을 적절하게 조절하여 해결할 수 있다.
    
   - 과소적합
   
       모델이 너무 단순해서 내재된 구조를 학습하지 못할 때를 의미하며 결국 훈련 데이터에서의 성능도 좋지 못하다는 것을 의미한다. 이를 해결하기 위해서는 좀 더 강력한 모델을 선택하거나, feature engineering을 거쳐 더 좋은 feature 제공 등이 있다.
       

#### 테스트와 검증

- 제공된 훈련 데이터를 train과 test용도로 나누어서 학습한 모델이 test 데이터에서의 일반화 성능을 파악해 **일반화 오차에 대한 추정값을 얻을 수 있다.**

- 결론적으로 일반화 성능에서 가장 중요한 것은 **Unseen data(한번도 보지 못한 데이터)**에서의 성능이 얼마나 좋게 나오는지이다 이를 꼭 명심하자.

   일반화 성능을 높이기 위한 방법으로는 홀드아웃 검증 -> 훈련데이터에서 검증 데이터를 따로 빼서 검증하는 방법. test 데이터는 그대로 존재. 
   
   교차 검증(Cross Validation)의 방법도 존재. -> 추후 자세히 설명.
   

- 데이터 불일치의 문제

    모델이 과대적합 된 것인지, 데이터의 불일치 때문인지 모르는 경우가 종종 발생하게 되는데 이때 훈련 데이터에서 일부를 떼어내어 **훈련 개발 세트(train-dev세트)를 만들고**, 훈련 데이터(train-dev 세트를 제외한)로 학습한 뒤 평가하면 데이터의 불일치 문제인지 아닌지를 판단할 수 있다.
